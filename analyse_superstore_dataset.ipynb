```python
# Calculate basic descriptive statistics for key metrics such as total sales, average order value, etc.
total_sales = df2["Quantity"] * df2["Sales"]
total_sale = total_sales.sum().astype(int)
print(f"Total Sales: {total_sale}")
```

```python
num_orders = df['Quantity'].count()
num_orders
average_order_value = total_sale / num_orders
average_order_value
print(f"Average Order Value: {average_order_value}")
```

```python
total_sales = df2["Quantity"] * df2["Sales"]
total_sales
```

```python
import matplotlib.pyplot as plt

total_sales.plot(kind='hist', bins=5, color='skyblue', edgecolor='black')

# Add labels and title
plt.xlabel('Total Sales')
plt.ylabel('Frequency')
plt.title('Distribution of Total Sales')
plt.grid(axis='y', alpha=0.75)

# Show plot
plt.show()
```

```python
# Plot the distribution of OrderQuantity using a histogram
df['Quantity'].plot(kind='hist', bins=5, color='lightgreen', edgecolor='black')

# Add labels and title
plt.xlabel('Order Quantity')
plt.ylabel('Frequency')
plt.title('Distribution of Order Quantity')
plt.grid(axis='y', alpha=0.75)

# Show plot
plt.show()
```

```python
# Identify top-selling products
top_products = df.groupby('Sub-Category')['Quantity'].sum().sort_values(ascending=False).head(5)

# Identify top-selling categories
top_categories = df.groupby('Category')['Quantity'].sum().sort_values(ascending=False).head(5)

print("Top Selling Products:")
print(top_products)

print("\nTop Selling Categories:")
print(top_categories)
```

```python
import numpy as np
# Define a date range
start_date = pd.to_datetime('2024-01-01')
end_date = pd.to_datetime('2024-05-24')

# Generate a list of all dates within the date range
date_range = pd.date_range(start_date, end_date, freq='D')

# Randomly assign dates to the DataFrame
df2['OrderDate'] = np.random.choice(date_range, size=len(df2))

print(df2)
```
